## Librer√≠as

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

## Carga del Dataset

```python
df = pd.read_csv("../data/raw/covid2020.csv")
df.head()
```

## Preprocesamiento

```python
#Cogemos las columnas que nos interesan.
df = df.drop(['Province/State','Lat', 'Long'], axis = 1)
df = df[["Country/Region", "12/31/20"]]
```

```python
#Cogemos los pa√≠ses que nos interesan.
listaPaisesEscogidos = ['Argentina','Brazil','Brunei','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','US']
df = df[df['Country/Region'].isin(listaPaisesEscogidos)].reset_index(drop=True)
```

```python
#Vemos que filas est√°n repetidas (mismo pa√≠s) y juntamos las filas del mismo pa√≠s, sumando sus datos.
df_completo = df.groupby(df['Country/Region']).aggregate(np.sum).reset_index()
df_completo = df_completo.rename(columns={'Country/Region': 'Country', '12/31/20': 'Covid_Deaths_2020'})
```

```python
df_completo['Country'] = ['Argentina','Brazil','Brunei Darussalam','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','United States of America']
```

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
pib_data = pd.read_csv('../data/raw/pib.csv',na_values='?')
```

```python
display(pib_data)
```

```python
paises = ['Argentina','Brazil','Brunei Darussalam','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','United States']
```

```python
pib_data.dropna().head(200)
```

```python
# quito filas
pib_data = pib_data.where(pib_data['Country Name'].isin(paises)).dropna()
pib_data
```

```python
# renobro las coumnas
new_col_names = [];
for name in pib_data.columns:
    new_col_names.append(name.split(" [")[0])
pib_data.columns = new_col_names
```

```python
# elimino columnas
before_1990 = pib_data.columns[4:34];
after_2017 = ["2018","2018"]

extra_fields = ['Series Name','Series Code', 'Country Code','2020']

pib_data = pib_data.drop(before_1990,axis = 1).drop(extra_fields,axis = 1)
```

```python
# camibio el nombr de russia y los emiratos arabes
pib_data['Country Name'] =  ['Argentina','Brazil','Brunei Darussalam','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','United States of America']
# reindexo el dataset
pib_data = pib_data.reset_index(drop=True)
```

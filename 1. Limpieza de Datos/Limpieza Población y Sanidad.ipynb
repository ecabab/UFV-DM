```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
population_data = pd.read_csv("../data/raw/Population_data.csv")
```

```python
#Seleccion de subconjunto de datos de estudio
Lista_paises =['Argentina','Brazil','Brunei Darussalam','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','United States']
population_selected = population_data.loc[population_data['Country Name'].isin(Lista_paises)]
population_selected.reset_index(drop = True,inplace= True)
```

```python
#Drops de columnas vacias o inservibles
before_1995 = population_selected.columns[range(4,28)]
after_2017 = population_selected.columns[range(51,len( population_selected.columns))]

population_first = population_selected.iloc[:,2:]
population_clean = population_first.drop(['Country Code'],axis = 1).drop(before_1995,axis = 1).drop(after_2017,axis = 1)
```

```python
#Modificación nombres columnas para cumplir un estandar
dic = {population_clean.columns.to_list()[0]:population_clean.columns.to_list()[0]}
columnas_nombres = population_clean.columns.to_list()
for i in range(1,len(columnas_nombres)):
    nombre = columnas_nombres[i]
    dic[columnas_nombres[i]] = str(nombre[:4])
population_clean.rename(columns = dic,inplace = True)
```

```python
population_clean['Country Name'] =  ['Argentina','Brazil','Brunei Darussalam','Canada','Colombia','Croatia','Ecuador','Finland','Germany','Hungary','Israel','Japan','Kuwait','Mexico','Nicaragua','Singapore','Slovenia','Spain','Sweden','United States of America']
```

```python
#Creacion columna sanidad
Sanidad = [0,0,1,1,0,0,0,1,1,0,1,1,1,0,0,1,1,1,1,0]
population_clean['Sanidad'] = Sanidad
```

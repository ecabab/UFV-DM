```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='whitegrid')
```

# Importación de datos

```python
# Indicamos la columna "Country Name" como índice
df_deaths = pd.read_csv('../../data/clean/deaths_clean.csv', index_col=1)
df_population = pd.read_csv('../../data/clean/population_clean.csv', index_col=1)
df_pib = pd.read_csv('../../data/clean/pib.csv', index_col=0)

# Borramos la columna de índices (números del 0 al 19)
df_deaths.drop('Unnamed: 0', axis=1, inplace=True)
df_population.drop('Unnamed: 0', axis=1, inplace=True)

# Transposing dataframes
df_deaths = df_deaths.transpose()
df_population = df_population.transpose()

# Exportando datos de sanidad
df_sanidad = df_population.loc['Sanidad']
df_population.drop('Sanidad', inplace=True)

# Convertimos el indice del PIB a numérico
df_pib.index = pd.to_numeric(df_pib.index)

# Convertimos el resto de valores del PIB a numéricos
df_pib = df_pib.apply(pd.to_numeric)
```

# Visualización de datos
## Muertes por año

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=df_deaths, dashes=False)
```

## Población por año

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=df_population, dashes=False)
```

## PIB por año

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=df_pib, dashes=False)
```

```python
# Pie chart
labels = ['Sanidad Privada', 'Sanidad Pública']
s_pub = df_sanidad.sum()
s_priv = len(df_sanidad) - s_pub
sizes = [s_priv, s_pub]

# Colores
colors = ['#ff9999','#66b3ff']

fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.show()
```

## Muertes por cada 100.000 habitantes por año

```python
df_deaths_ratio = df_deaths/df_population*100000
```

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=df_deaths_ratio, dashes=False)
```

## PIB per cápita

```python
df_population.index = df_population.index.astype('int64')
df_pib_pcap = df_pib/df_population
```

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=df_pib_pcap, dashes=False)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import cm as cm
import numpy as np
```

# Importaci칩n de datos

```python
df_pca = pd.read_csv('../../data/final/data_rows.csv')
```

# Preparaci칩n de datos

```python
df_pca.drop(columns=['Unnamed: 0'],axis = 1,inplace = True)
columnas = ['deaths','populantion','pib']
df_pca_clean = df_pca[columnas]
df_pca_clean.head()
```

```python
scaler = StandardScaler()
df_pca_scaled = scaler.fit_transform(df_pca_clean.values)
```

# Implementaci칩n PCA
## PCA con target Pais

```python
# PCA de 2 componentes
pca = PCA(n_components=2)

# Calculamos los componentes principales, pasando como par치metro los datos normalizados
principalComponents = pca.fit_transform(df_pca_scaled)

# Creamos un dataset con los componentes principales del PCA
components_df = pd.DataFrame(data = principalComponents, columns = ['Componente 1', 'Componente 2'])
components_df['target'] = df_pca['Country Name']
```

```python
components_df.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente 1', fontsize = 15)
ax.set_ylabel('Componente 2', fontsize = 15)
ax.set_title('PCA de 2 Componentes', fontsize = 20)
colors = cm.get_cmap('Paired',20)
colors  = [colors(i) for i in range(20)]
targets = ['Argentina', 'Brazil', 'Brunei Darussalam', 'Canada', 'Colombia',
       'Croatia', 'Ecuador', 'Finland', 'Germany', 'Hungary', 'Israel',
       'Japan', 'Kuwait', 'Mexico', 'Nicaragua', 'Singapore', 'Slovenia',
       'Spain', 'Sweden', 'United States of America']

for target, color in zip(targets,colors):
    idxToKeep = components_df['target'] == target
    ax.scatter(components_df.loc[idxToKeep, 'Componente 1']
               , components_df.loc[idxToKeep, 'Componente 2']
               , color = color
               , s = 20)

ax.legend(targets)
ax.grid()
```

## Exploraci칩n de la zona izquierda

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente 1', fontsize = 15)
ax.set_ylabel('Componente 2', fontsize = 15)
ax.set_title('PCA de 2 Componentes', fontsize = 20)
colors = cm.get_cmap('Paired',19)
colors  = [colors(i) for i in range(19)]
targets = ['Argentina', 'Brazil', 'Brunei Darussalam', 'Canada', 'Colombia',
       'Croatia', 'Ecuador', 'Finland', 'Germany', 'Hungary', 'Israel',
       'Japan', 'Kuwait', 'Mexico', 'Nicaragua', 'Singapore', 'Slovenia',
       'Spain', 'Sweden', 'United States of America']

for target, color in zip(targets,colors):
    idxToKeep = components_df['target'] == target
    ax.scatter(components_df.loc[idxToKeep, 'Componente 1']
               , components_df.loc[idxToKeep, 'Componente 2']
               , color = color
               , s = 20)

ax.legend(targets)
ax.grid()
```

## PCA con target Sanidad

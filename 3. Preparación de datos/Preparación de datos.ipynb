```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
population_clean = pd.read_csv("../data/clean/population_clean.csv")
pib_clean = pd.read_csv("../data/clean/pib_clean.csv")
deaths_clean = pd.read_csv("../data/clean/deaths_clean.csv")
covid2020_clean = pd.read_csv("../data/clean/covid_clean.csv")
```

```python
deaths_final = deaths_clean.merge(population_clean, how="inner", on='Country Name',  suffixes=('_death', '_population')).drop(['Unnamed: 0_population','Unnamed: 0_death'],axis = 1)
deaths_final = deaths_final.merge(pib_clean.add_suffix('_pib'), how="inner", left_on='Country Name', right_on='Country Name_pib').drop('Unnamed: 0_pib',axis = 1)
```

```python
display(deaths_final)
#deaths_final.to_csv("../data/final/deaths_final.csv")
```

```python
new = []
for index, row in deaths_final.iterrows():
    for year in range(1995,2018):
        new_row = {'Country Name':row['Country Name'], 'Sanidad':row['Sanidad'], 'year':year,'deaths':row[str(year)+'_death'], 'populantion': row[str(year)+'_population'],'pib': row[str(year)+'_pib']};
        new.append(new_row)
deaths_rows_final = pd.DataFrame(new)
```

```python
display(deaths_rows_final)
#deaths_rows_final.to_csv("../data/final/deaths_rows_final.csv")
```
